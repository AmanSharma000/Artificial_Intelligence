```python
import numpy as np
import pandas as pd 
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(y_train.shape,y_test.shape)
```

```python
SGD = SGDRegressor()
```

```python
SGD.fit(X_train,y_train)
```

```python
y_pred=SGD.predict(X_test)
```

```python
print(SGD.intercept_)
print(SGD.coef_)
```

```python
r2_score(y_test,y_pred)
```

```python
class SGDR:

    def __init__(self, learning_rate=0.01, epochs=100):
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):

            indices = np.random.permutation(X_train.shape[0]) # indices = [120, 5, 301, 88, ..., 17]   # total 353 numbers

            for idx in indices: #due to this same sample does not repeat

                y_hat = np.dot(X_train[idx], self.coef_) + self.intercept_
                intercept_der = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                
                coef_der = -2 * (y_train[idx] - y_hat) * X_train[idx]
                self.coef_ = self.coef_ - (self.lr * coef_der)

    def predict(self, X_test):
        return np.dot(X_test, self.coef_) + self.intercept_
```

```python
SGDR = SGDR()
```

```python
SGDR.fit(X_train,y_train)
```

```python
y_pred=SGDR.predict(X_test)
```

```python
print(SGDR.intercept_)
print(SGDR.coef_)
```

```python
r2_score(y_test,y_pred)
```
