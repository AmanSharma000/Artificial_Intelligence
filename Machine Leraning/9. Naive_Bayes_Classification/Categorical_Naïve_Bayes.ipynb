```python
df.drop(columns=['day'], inplace=True)
```

```python
x=df.iloc[:,0:4]
```

```python
x.head()
```

```python
y=df.iloc[:,4]
```

```python
y
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
X_encoded = encoder.fit_transform(x)
```

```python
X_encoded
```

```python
from sklearn.naive_bayes import CategoricalNB

model = CategoricalNB()
model.fit(X_encoded, y)
```

```python
sample = pd.DataFrame([{
    "outlook": "Sunny",
    "temp": "Cool",
    "humidity": "High",
    "wind": "Weak"
}])


sample_encoded = encoder.transform(sample)
print(model.predict(sample_encoded))
print(model.predict_proba(sample_encoded))
```
