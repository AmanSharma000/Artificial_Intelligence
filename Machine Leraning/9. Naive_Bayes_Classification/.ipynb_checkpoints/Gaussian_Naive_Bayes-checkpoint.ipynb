```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=['Geography', 'Gender'],
    drop_first=True   # k-1 encoding
)


```

```python
df_encoded.head()
```

```python
X=df_encoded.drop(columns=['Exited'])
```

```python
X.head()
```

```python
y=df_encoded['Exited']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=36,stratify=y)
```

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pre=model.predict(X_test)
```

```python
y_pre[12]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,confusion_matrix,roc_auc_score,roc_curve
```

```python
accuracy = accuracy_score(y_test, y_pre)
precision = precision_score(y_test, y_pre)
recall = recall_score(y_test, y_pre)


print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
confusion_matrix(y_test, y_pre)
```

```python
y_prob=model.predict_proba(X_test)[:, 1]
```

```python
y_prob        #first - > 0
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
```

```python
youden_j = tpr - fpr
print(youden_j)
best_idx = np.argmax(youden_j)
print(best_idx)
best_threshold = thresholds[best_idx]

```

```python
print("Best threshold (ROC):", best_threshold)
```

```python
y_pred_best = (y_prob >= best_threshold).astype(int)
```

```python
from sklearn.metrics import classification_report
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print(confusion_matrix(y_test, y_pred_best))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(fpr, tpr, label=f"ROC (AUC = {roc_auc_score(y_test, y_prob):.3f})")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.scatter(fpr[best_idx], tpr[best_idx], color='red',
            label=f"Best threshold = {best_threshold:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
