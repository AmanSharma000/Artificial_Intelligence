```python
import numpy as np
import pandas as pd
import time
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR,LinearSVR
from sklearn.kernel_approximation import RBFSampler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x,y=fetch_california_housing(return_X_y=True)
```

```python
x.shape
```

```python
y.shape
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    x, 
    y,  
    test_size=0.25,
    random_state=42,
)
```

```python
# Standardize
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print(housing.data.shape, housing.target.shape)

print(housing.feature_names[0:6])
```

```python
start = time.time()
lin_svr = LinearSVR()
lin_svr.fit(X_train, y_train)
svr_lin_time = time.time() - start
y_pred_lin_svr=lin_svr.predict(X_test)

```

```python
svr_lin_time
```

```python
print(mean_squared_error(y_test,y_pred_lin_svr), r2_score(y_test,y_pred_lin_svr))
```

```python
#--------------------------
```

```python
start = time.time()
rbf_svr = SVR(kernel="rbf",gamma="scale")
rbf_svr.fit(X_train, y_train)
svr_rbf_time = time.time() - start
y_pred_rbf_svr=rbf_svr.predict(X_test)

```

```python
svr_rbf_time
```

```python
print(mean_squared_error(y_test,y_pred_rbf_svr), r2_score(y_test,y_pred_rbf_svr))
```
