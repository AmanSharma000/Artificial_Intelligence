```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
df=pd.read_csv("tv_sales.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
#TV : The amount spent on TV advertising (in thousands of dollars)
#Sales : The resulting sales volume (in thousands of units)
```

```python
plt.scatter(df['TV'],df['Sales'])
plt.xlabel('TV:spent amount')
plt.ylabel('Sales:resulting sales volume')
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
print(x.shape,y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2) 
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
sml=LinearRegression()
```

```python
sml.fit(x_train,y_train)
```

```python
y_pred=sml.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
```

```python
mae
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
r2
```

```python
adj_r2 = 1 - ((1-r2)*(200-1)/(200-1-1))
```

```python
adj_r2
```

```python
m=sml.coef_[0]
```

```python
m
```

```python
b=sml.intercept_
```

```python
b
```

```python
plt.scatter(df['TV'],df['Sales'])
plt.plot(x_train,sml.predict(x_train),color='red')
plt.xlabel('TV:spent amount')
plt.ylabel('Sales:resulting sales volume')
```
